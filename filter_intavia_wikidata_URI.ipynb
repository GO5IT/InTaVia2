## Create a list of Wikidata URIs found in InTaVia

```python
URI_list_df  = pd.read_csv(filelocation_local + 'PersonURIs_CSV/merge_at_nl_fi_si.csv', index_col=0)
intavia_wikidataURI_list = (URI_list_df['wikidataURI']).to_list()
#print(intavia_wikidataURI_list)
# Take Q-prefix labels after last slash
intavia_wikidataURI_list = [wikidataURI.rsplit('/', 1)[1] for wikidataURI in intavia_wikidataURI_list]
print(intavia_wikidataURI_list)
```

## Create a list of files in the EuropeanaObjectURIlist folder

```python
# folder path to CSV files (Uncomment the path to be used for Creator and Contributor)
dir_path = filelocation_local + 'EuropeanaObjectURIlist/Creator/'
#dir_path = filelocation_local + 'EuropeanaObjectURIlist/Contributor/'
print(dir_path)
# Create an empty list
list_file = []
# Iterate in the folder
for path in os.listdir(dir_path + 'all_wikidataURI/'):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path + 'all_wikidataURI/', path)):
        # check file is csv file
        if '.csv' in path:
            list_file.append(path)
        else:
            pass
list_file.sort()
print(list_file)
```

```python
# Uncomment the next 2 lines, when testing with only two files
#list_file = list_file[1:3]
#print(list_file)

# Iterate over CSV files
i = 0
# Empty list to compile number of items processed
iilist = [] 
for n in list_file:
  print(n + ' is started as file no ' + str(i))
  URI_list_df  = pd.read_csv(dir_path + 'all_wikidataURI/' + n)

  # Iterate over URI list
  # Create an empty dataframe by pandas
  new_URI_list_df = pd.DataFrame(columns=['Europeana_proxy', 'External_URI','Wikidata_URI'])
  ii = 0
  for index, row in URI_list_df.iterrows():
    item = row['Wikidata_URI'].rsplit('/', 1)[1]

    if item in intavia_wikidataURI_list:
      dicts = {'Europeana_proxy': row['Europeana_proxy'], 'External_URI': row['External_URI'], 'Wikidata_URI': row['Wikidata_URI']}
      new_URI_list_df_dicts = pd.DataFrame([dicts])
      new_URI_list_df = pd.concat([new_URI_list_df, new_URI_list_df_dicts], ignore_index=True)
      ii = ii + 1    
    else:
      pass
      #print('No match')
    #print('Processing ' + str(item) + ' completed')
    #print('-----------------')  
  #print(new_URI_list_df)

  # #Saving as CSV file
  df_csv = new_URI_list_df.to_csv(dir_path + 'OnlyInTaVia_at_nl_fi_si_' + n)
  #print('-----------------')
  #print(n + ' is finished, thus ' + str(ii) + ' items are processed')
  i = i + 1
  iilist += [ii]
print('The numbers of URIs included per source CSV file are: ' + str(iilist))
sum_items = sum(iilist)
print('Total of ' + str(sum_items) + ' items were processed')
```
